```python
def check_strike(S):
    S = S.lower()  
    x_strike = False
    o_strike = False
    count = 1
    
    for i in range(1, len(S)):
        if S[i] == S[i - 1]:
            count += 1
        else:
            count = 1

        if count >= 3:
            if S[i] == 'x':
                x_strike = True
            elif S[i] == "o":
                o_strike = True

            if x_strike and o_strike:
                break

    if x_strike and o_strike:
        return 'tie'
    elif x_strike:
        return 'x'
    elif o_strike:
        return 'o'
    else:
        return 'tie'
    
print(check_strike("ooxoxxoooxxx"))
```
